```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

compravendita = pd.read_csv("Compravendita_Pesticidi.csv")

pesticidi = sns.barplot(data = compravendita, x = "Stati", y = "Valore", hue = "Anni", errorbar = None, palette= "pastel")
sns.move_legend(obj=pesticidi, loc='upper right', bbox_to_anchor=(1.25, 1))

pesticidi.set_ylim(0,90000)
pesticidi.set_xlabel(None)
pesticidi.set_ylabel(None)

sns.set_style(rc={'axes.facecolor':'#efefef','figure.facecolor': '#efefef'})

plt.text(
    0, -0.1,  # Posizione
    "Fonte: Eurostat",
    fontsize=10,
    color='gray',
    ha='left',  # Allineamento orizzontale
    transform=plt.gca().transAxes  # Posiziona il testo in coordinate relative al grafico
)
```
