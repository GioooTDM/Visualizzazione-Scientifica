```python
import numpy as np
import pandas as pd
import seaborn as sea

gdp_file = '..\data\gdp_perCapita.csv'
ColorStates= {'Italy':'green', 'Spain':'red', 'Hungary': "purple", 'Ireland': 'orange', 'Sweden':'blue'}
df = pd.read_csv(gdp_file, sep=',', usecols=['geo', 'TIME_PERIOD', 'OBS_VALUE']).rename(columns={
    "geo":"state", "TIME_PERIOD": "year", "OBS_VALUE":"GDP" })
dfPivoted = df.pivot(index='year', columns='state', values='GDP')
highlight = {state:ColorStates[state] if state in ColorStates.keys() else 'lightgray' for state in df['state']}
sea.set_theme(style="white")
lnPlot=sea.lineplot(data=dfPivoted, legend=True, palette=highlight, linewidth=2.5)
sea.move_legend(lnPlot,'upper right', bbox_to_anchor=(1.5, 1))
```
